```python
import cv2              
import numpy as np      
import matplotlib.pyplot as plt  
```

```python
img_path = 'images/01.jpg'

frame = cv2.imread(img_path)

# ---------------------------------------
# Convert image to grayscale
# ---------------------------------------
gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)


# ---------------------------------------
# Step 2: Detect Circles using Hough Transform
# ---------------------------------------Q

circles = cv2.HoughCircles(
    gray,                   # Input image (must be grayscale)

    cv2.HOUGH_GRADIENT,     # Detection method using gradient information

    dp=1.1,                 # Inverse ratio of accumulator resolution
                            # 1.1  accumulator slightly smaller than image
Q
    minDist=30,             # Minimum distance between detected circle centers
                            # Prevents detecting multiple circles close together

    param1=120,             # Upper threshold for Canny edge detection
                            # Higher value  stronger edges only

    param2=30,              # Threshold for center detection
                            # Lower value  more circles detected
                            # Higher value  fewer, more confident circles

    minRadius=40,           # Smallest circle radius allowed (iris minimum)

    maxRadius=90            # Largest circle radius allowed (iris maximum)
)


# ---------------------------------------
# Step 3: Draw Detected Circles
# ---------------------------------------

# Check if any circles were detected
if circles is not None:

    # Round values and convert to unsigned integers
    circles = np.uint16(np.around(circles))

    # Loop through all detected circles
    for c in circles[0, :]:
        x, y, r = c          # x  center x, y  center y, r  radius

        cv2.circle(frame, (x, y), r, (255, 0, 0), 2)

        # Draw the center
        cv2.circle(frame, (x, y), 4, (0, 0, 255), -1)


frame = cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10,10))   # Increase figure size (width, height in inches)
plt.imshow(frame)
plt.title('Circles Detection')
plt.show()
```

```python

img_path = 'images/04.jpg'

frame = cv2.imread(img_path)

# ---------------------------------------
# Convert image to grayscale
# ---------------------------------------
gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

# ---------------------------------------
# Reduce noise in grayscale image
# ---------------------------------------

# cv2.medianBlur  removes salt-and-pepper noise
# gray  input grayscale image
# 5  kernel size (must be odd)
gray = cv2.medianBlur(gray, 5)

# ---------------------------------------
# Step 2: Detect Circles using Hough Transform
# ---------------------------------------

circles = cv2.HoughCircles(
    gray,                   # Input image (must be grayscale)

    cv2.HOUGH_GRADIENT,     # Detection method using gradient information

    dp=1.1,                 # Inverse ratio of accumulator resolution
                            # 1.1  accumulator slightly smaller than image

    minDist=60,             # Minimum distance between detected circle centers
                            # Prevents detecting multiple circles close together

    param1=125,             # UApper threshold for Canny edge detection
                            # Higher value  stronger edges only

    param2=30,              # Threshold for center detection
                            # Lower value  more circles detected
                            # Higher value  fewer, more confident circles

    minRadius=30,           # Smallest circle radius allowed (iris minimum)

    maxRadius=100           # Largest circle radius allowed (iris maximum)
)


# ---------------------------------------
# Step 3: Draw Detected Circles
# ---------------------------------------

# Check if any circles were detected
if circles is not None:

    # Round values and convert to unsigned integers
    circles = np.uint16(np.around(circles))

    # Loop through all detected circles
    for c in circles[0, :]:
        x, y, r = c          # x  center x, y  center y, r  radius

        cv2.circle(frame, (x, y), r, (255, 0, 0), 2)

        # Draw the center
        cv2.circle(frame, (x, y), 4, (0, 0, 255), -1)


# ---------------------------------------
# Display the Result
# ---------------------------------------

frame = cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10,10))   # Increase figure size (width, height in inches)
plt.imshow(frame)
plt.title('Circles Detection')
plt.show()
```
